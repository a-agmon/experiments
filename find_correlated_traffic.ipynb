```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def unit_vector(vector):
    return vector / np.linalg.norm(vector)

def angle_between(v1, v2):
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
```

```python
v1 = np.array([16282,15876,18380,18772,16828,18147,
               16910,17853,16517,21265,10040,1510,
               569,525,927,478,844,1086,1390,1541,
               1557,961,1051,1543,1399,2056,6993,4746,3142])

v2 = np.array ([7723,7308,9018,10325,10317,10826,9631,
                7486,7971,11398,6433,1082,361,217,334,
                198,394,442,519,588,627,346,435,609,509,960,2497,1760,1283])

v3 = np.array ([3000,3500,3500,4500,5600,5000,4500,
                4500,4500,2500,5000,4500,4000,3500,3000,
                5000,2600,4000,6000,3000,5000,3000,3000,3000,3000,3000,2497,1760,1283])
```

```python
plt.plot(v1, marker='', color='grey', linewidth=2, label='v1')
plt.plot(v2, marker='', color='blue', linewidth=2 , label='v2', alpha=0.6)
plt.plot(v3, marker='.', color='red', linewidth=2,linestyle='dashed', label='v3', alpha=0.5)
plt.legend()
```

```python
a1 = angle_between(v1,v2)
print(f'angle beteen correlated lines:\n{a1}')
```

```python
a2 = angle_between(v1,v3)
a3 = angle_between(v2,v3)
print(f'angle beteen NOT correlated lines:\n{a2}\n{a3}')
```
